```python
ba_graph = nx.generators.random_graphs.barabasi_albert_graph(100, 10)
G = ig.Graph.from_networkx(ba_graph)
communities = leidenalg.find_partition(G, leidenalg.ModularityVertexPartition)

# visualize
nx.draw(ba_graph, 
        pos=nx.spring_layout(ba_graph),
        node_color=communities.membership, 
        labels=dict(zip(ba_graph.nodes, communities.membership)))
```
